```python
df['Time_stamp'] = df['Title'].str.extract(r'([A-Za-z]{3} \d{4} - [A-Za-z]{3} \d{4})')
```

**Calcuate duration from time stamp**

```python
def duration_in_days(date_range: str) -> int:
    # Split the range
    start_str, end_str = date_range.split(" - ")

    # Convert to datetime (using first day of the month)
    start_date = pd.to_datetime(start_str, format="%b %Y")
    end_date = pd.to_datetime(end_str, format="%b %Y")

    # Calculate difference
    duration = (end_date - start_date).days

    return duration
#print(calculate_duration("Apr 2009 - Jul 2010"))
```

```python
df['Duration_days'] = df['Time_stamp'].apply(duration_in_days)
```
